```python
import numpy as np
import matplotlib.pyplot as plt

n_samples= 1000
```

```python
X = 2*np.random.rand(n_samples,2)-1
y = X[:,1]>=0.5*np.sin(np.pi*X[:,0])


plt.scatter(X[:, 0], X[:, 1], c=y)
plt.axis("scaled")
```

```python
X = 2*np.random.rand(n_samples,2)-1
y = X[:,1]>=(X[:,0])**2-0.5


plt.scatter(X[:, 0], X[:, 1], c=y)
plt.axis("scaled")
```

```python
X = 2*np.random.rand(n_samples,2)-1
y = X[:,1]>=(X[:,0])**2-0.5

# Parameter
theta = np.linspace(0, 2*np.pi, 1000)

# Radial function (shifted to avoid negative radius)
r = 1 + 0.5 * np.cos(5 * theta)

# Convert to Cartesian
xx = r * np.cos(theta)
yy = r * np.sin(theta)

ang = np.atan2(X[:,1],X[:,0])
rad = np.sqrt(X[:,1]**2+X[:,0]**2)

y = rad*(1+0.5* np.cos(5*ang))>=0.5

# Plot
plt.figure(figsize=(6,6))
plt.plot(xx, yy)
plt.gca().set_aspect('equal')
plt.title("5-pointed star (radial parametrization)")
plt.grid(True)
plt.show()

plt.scatter(X[:, 0], X[:, 1], c=y)
plt.axis("scaled")
```
