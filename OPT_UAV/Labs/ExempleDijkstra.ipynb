1. Project Euler, Problem 18: https://projecteuler.net/problem=18
- No need to create graph: the source node is the top, unvisited neighbors are always the two numbers directly below. A dynamical programming solution can also be given, but may be more costly.

  

```python
import numpy as np

A = [[3,0,0,0],
     [7,4,0,0],
     [2,4,6,0],
     [8,5,9,3]]

A = np.array(A)

print(A)
```

Remarks:
- nodes below value on position $i$ have indices $i$ and $i+1$
- we will keep a matrix having the same size as `A` and we'll update the value of a node if a **bigger** value is found
- 

```python
nrows = A.shape[0]  # number of rows in A

Cost = np.zeros(A.shape)

Cost[0][0] = A[0][0]

print(A)

for i in range(1,nrows):
    for j in range(0,i):
        #print(i," ",j," ",A[i][j])
        vleft = Cost[i-1][j]+A[i][j]
        vright = Cost[i-1][j]+A[i][j+1]
        if vleft>Cost[i,j]:
            Cost[i,j] = vleft
        if vright>Cost[i,j+1]:
            Cost[i,j+1] = vright
    print(Cost)
```

```python
def maxPyramid(A):
    nrows = A.shape[0]  # number of rows in A

    Cost = np.zeros(A.shape)

    Cost[0][0] = A[0][0]

    print(A)

    prev = np.zeros(A.shape)
    print(prev)

    for i in range(1,nrows):
        for j in range(0,i):
            #print(i," ",j," ",A[i][j])
            vleft = Cost[i-1][j]+A[i][j]
            vright = Cost[i-1][j]+A[i][j+1]
            if vleft>Cost[i,j]:
                Cost[i,j] = vleft
                prev[i,j] = j
            if vright>Cost[i,j+1]:
                Cost[i,j+1] = vright
                prev[i,j+1] = j
    return max(Cost[-1]), prev, Cost[-1]

maxPyramid(A)
```

```python
def readTriMat(s):
    L = []
    with open(s) as file:
        for line in file:
            L.append(np.fromstring(line,sep=" "))
    #print(L)

    BigA = np.zeros((len(L),len(L)))
    for i in range(0,len(L)):
        #print(L[i])
        #print(BigA[i][0:i+1])
        BigA[i][0:i+1] = L[i]
    return (BigA)
    
BigA = readTriMat('0067_triangle.txt')

maxPyramid(BigA)
```

2. Project Euler, problem 81: https://projecteuler.net/problem=81   
- there is no need to create a graph, just apply Dijkstra's algorithm on the matrix, observing the unvisited neighbors of a current element

```python
A = np.array([[131 ,673, 234, 103, 18],
             [201,96,342,965,150],
             [630,803,746,422,111],
             [537,699,497,121,956],
             [805,732,524,37,331]])

print(A)
```

To explore the matrix we go either right or down. Start from the top left node and visit all elements. Replace value in cost matrix if it is smaller than before.



```python
# initialize with infinity since we are looking for shortest paths

Cost = np.matrix(np.ones(A.shape) * np.inf)
print(Cost)

queue = [(0,0)]
nmax = len(Cost)

Cost[0,0] = A[0,0]

while len(queue)>0:
    # pop the first element
    q = queue.pop(0);
    i = q[0]
    j = q[1]
    
    # down 
    if (i+1)<nmax:
        nextl = (i+1,j)
        queue.append(nextl)
        nextcost = Cost[i,j]+A[i+1,j]
        if nextcost<Cost[i+1,j]:
            Cost[i+1,j] = nextcost
    # left
    if (j+1)<nmax:
        nextl = (i,j+1)
        queue.append(nextl)
        nextcost = Cost[i,j]+A[i,j+1]
        if nextcost<Cost[i,j+1]:
            Cost[i,j+1] = nextcost

print(Cost)
```
