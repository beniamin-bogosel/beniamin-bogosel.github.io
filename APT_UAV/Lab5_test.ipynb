```python
# definirea unei liste
L = ["apple","banana","cherry"]
print(L)
for i in range(0,len(L)):
    print(L[i])
```

```python
# s-a copiat adresa si daca modificam L0 se modifica si L
# folosim copy ca sa duplicam obiectele
L0 = L.copy()

# adaugare in capatul listei
L0.append("orange")

print(L0)
print(L)
```

```python
# inserare in interiorul listei
L2 = L.copy()
print(L2)
L2.insert(0,"orange")
print(L2)
```

```python

print(L2)
# eliminare si returnare de pe ultima pozitie
x = L2.pop()
print(x)
print(L2)
```

```python
# eliminare si returnare de pe o pozitie data
print(L2)
x = L2.pop(0)
print(x)
print(L2)
```

```python
# parcurgem sirul, punem parantezele stangi intr-o lista 
# cand intalnim o paranteza dreapta, comparam cu ultima paranteza 
# si o scoatem din lista, daca aceasta corespunde
# daca nu: eroare

str = "[[{((a+b)/2+3)-5]}}"

# lista vida
L = []

# 1. exploram sirul caracter cu caracter
for i in range(0,len(str)):
    # parantezele stangi se adauga
    if str[i] in ("(","[","{"):
        L.append(str[i])
    
    # paranteze din dreapta
    if str[i]==")":
        try:
            paranteza = L.pop()
        except:
            print("Eroare: prea multe paranteze stangi")
        if paranteza != "(":
            print("Eroare")

    if str[i]=="]":
        try:
            paranteza = L.pop()
        except: 
            print("Eroare: prea multe paranteze stangi")

        if paranteza != "[":
            print("Eroare")

    if str[i]=="}":
        try:
            paranteza = L.pop()
        except: 
            print("Eroare: prea multe paranteze stangi")
        if paranteza != "{":
            print("Eroare")

if len(L)>0:
    print("Eroare: prea multe paranteze stangi")

```

```python
# Exercitiul 2: verificarea anagramelor

str1 = "romania"
str2 = "armonia"

# Liste caractere sortate
Lista1 = sorted(str1)
Lista2 = sorted(str2)

print(Lista1==Lista2)

# solutia sub forma de functie
def CheckAnagram(str1,str2):
    Lista1 = sorted(str1)
    Lista2 = sorted(str2)

    return Lista1 == Lista2

print(CheckAnagram("romania","armonia"))
```
